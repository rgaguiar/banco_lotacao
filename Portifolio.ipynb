```python
import pandas as pd
import pandasql as ps
```

### Lendo as bases de dados

```python
tb_cargo = pd.read_csv('https://raw.githubusercontent.com/rgaguiar/banco_lotacao/main/cargo.csv')
tb_departamento = pd.read_csv('https://raw.githubusercontent.com/rgaguiar/banco_lotacao/main/departamento.csv')
tb_funcionario = pd.read_csv('https://raw.githubusercontent.com/rgaguiar/banco_lotacao/main/funcionario.csv')
tb_lotacao = pd.read_csv('https://raw.githubusercontent.com/rgaguiar/banco_lotacao/main/lotacao.csv')
```

#### Problema de negócio
1. Liste nome (funcionário e cargo) por departamento com salário e pretensão salarial.

```python
query = """
select
fc.nome as nm_funcionario,
cg.nome as nm_cargo,
fc.pretensao_salarial,
lt.salario
from tb_lotacao as lt
inner join tb_funcionario as fc on fc.id = lt.id_funcionario
inner join tb_cargo as cg on cg.id = lt.id_cargo
"""

ps.sqldf(query, locals())
```

A consulta resulta na lista dos funcionarios por departamento com salario e pretensão salarial.

2. Listar os funcionários (nome e salário) que não sejam PCD e que possuam salário maior que R$ 4.000,00 ordenados pelo maior salário.

```python
print('Tabela Lotação')
display(tb_lotacao.head(2))
print('Tabela Funcionario')
display(tb_funcionario.head(2))
```

```python
query = """
select
fc.nome,
fc.pcd,
lt.salario
from tb_lotacao as lt
inner join tb_funcionario as fc on fc.id = lt.id_funcionario
where fc.pcd = 0 and lt.salario > 4000
order by lt.salario desc
"""

ps.sqldf(query, locals())
```

Lista dos funcionarios que não são pcd e com salario maior que R$4000

3. Liste os cargos e suas respectivas médias salariais maiores que R$5.000,00.


```python
print('Tabela Cargo')
display(tb_cargo.head(2))
print('Tabela Lotação')
display(tb_lotacao.head(2))
```

```python
query = """
select
cg.nome as cargo,
avg(lt.salario) as media_salario
from tb_lotacao as lt
inner join tb_cargo as cg on cg.id = lt.id_cargo
group by
cg.nome
having avg(lt.salario) > 5000
"""

ps.sqldf(query, locals())
```

LIsta dos cargos por media salarial maior que R$ 5000
